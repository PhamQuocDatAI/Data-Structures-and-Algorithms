```python
# Tháp Hà Nội
def THN(n, A, B, C):
    if n == 1:
        print("Di chuyển vòng 1 từ nguồn", A, "đến đích", B)
        return 
    THN(n-1, A, C, B)
    print("Di chuyển vòng", n, "từ nguồn", A, "đến đích", B)
    THN(n-1, A, B, C)

n = int(input("Số tầng của tháp: "))
THN(n, "A", "B", "C")  
```

```python
# Ước số chung lớn nhất
def UCLN(a, b):
    if (b == 0):
        return a;
    return UCLN(b, a % b);
 
a = int(input("Nhập số nguyên dương a = "));
b = int(input("Nhập số nguyên dương b = "));
#tính USCLN của a và b
print("Ước chung lớn nhất của", a, "và", b, "là:", UCLN(a, b))
```

```python
# Tính giai thừa của một số
def giai_thua(n):
    while n < 0:
        print("Lỗi giá trị n.")
        n = int(input("Nhập giá trị n: "))
    if n == 0:
        return 1
    return n * giai_thua(n-1)

n = int(input("Nhập giá trị n: "))
print(giai_thua(n))
```

2. Danh sách liên kết
    - Cài đặt danh sách liên kết đơn
    - Cài đặt danh sách liên kết kép
    - Cài đặt ngăn xếp - stack
    - Cài đặt hàng đợi - queue

```python
# Danh sách liên kết đơn
class Node:
  

    def __init__(self, data):
        self.data = data  
        self.next = None  
        
        
class LinkedList:
  

    def __init__(self):
        self.head = None
  
  
    def push(self, new_data):
  

        new_node = Node(new_data)
  
        new_node.next = self.head
  
        self.head = new_node
  

    def insertAfter(self, prev_node, new_data):
  
        if prev_node is None:
            print("The given previous node must inLinkedList.")
            return
  

        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node
  
    def append(self, new_data):

        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
  
        last = self.head
        while (last.next):
            last = last.next
  
        last.next =  new_node
  
    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data)
            temp = temp.next


if __name__=='__main__':
    llist = LinkedList()
    llist.append(6)
    llist.push(7);
    llist.push(1);
    llist.append(4)
    llist.insertAfter(llist.head.next, 8)
    print('Created linked list is:')
    llist.printList()
```

3. Cây
    - Cài đặt cây - duyệt cây theo thứ tự trước
    - Cài đặt cây - duyệt cây theo thứ tự sau

```python
import random 

class Node:
    
    
    def __init__(self, data):
        self.left_child = None
        self.right_child = None
        self.parent_node = data

    def print_tree(self):
        if self.left_child:
            self.left_child.print_tree()
        print(self.parent_node)
        if self.right_child:
            self.right_child.print_tree()

    def insert_number(self, input_number):

        if self.parent_node:

            if input_number < self.parent_node:
                if self.left_child is None:
                    self.left_child = Node(input_number)
                else:
                    self.left_child.insert_number(input_number)

            elif input_number > self.parent_node:
                if self.right_child is None:
                    self.right_child = Node(input_number)
                else:
                    self.right_child.insert_number(input_number)
        else:
            self.parent_node = input_number


def rd():
    return random.randint(-50, 50)
```

```python
root = Node(rd())
root.insert_number(rd())
root.insert_number(rd())
root.insert_number(rd())
root.insert_number(rd())
root.insert_number(rd())
root.insert_number(rd())

root.print_tree()
```

5. Sắp xếp và tìm kiếm
    - Cài đặt thuật toán sắp xếp chọn
    - Cài đặt thuật toán sắp xếp chèn
    - Cài đặt thuật toán sắp xếp nổi bọt
    - Cài đặt thuật toán sắp xếp nhanh - quick sort
    - Cài đặt thuật toán heap sort
    - Cài đặt thuật toán sắp xếp trộn - merge sort

```python
# Select Sort
import random 


A = []
for x in range(10):
    a = random.randint(-50, 50)
    A.append(a)
    
print('Mảng:',A) 
print('-'*50)
for i in range(len(A)):
    min_idx = i
    for j in range(i+1, len(A)):
        if A[min_idx] > A[j]:
            min_idx = j      
    A[i], A[min_idx] = A[min_idx], A[i]
      
print ("Sorted array")
for i in range(len(A)):
    print("%d" %A[i])
```

```python
# Insertion Sort
def insertionSort(arr):
  
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
        
        
B = []
for x in range(10):
    b = random.randint(-50, 50)
    B.append(b)
    
print("Array B:", B)
print('-'*50)
insertionSort(B)
print("Sorted array:")
for i in range(len(B)):
    print("%d" %B[i])
```

```python
# Bubble Sort
# Insertion Sort
def bubbleSort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
C = []
for x in range(10):
    c = random.randint(-50, 50)
    C.append(c)

print('Array C:', C)
print('-'*50)

bubbleSort(C)  
print ("Sorted array is:")
for i in range(len(C)):
    print ("%d" %C[i]), 
```

```python
# Quick Sort 

def partition(arr, low, high):
    i = (low-1)         
    pivot = arr[high]    
  
    for j in range(low, high):

        if arr[j] <= pivot:
  

            i = i+1
            arr[i], arr[j] = arr[j], arr[i]
  
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)

  
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:

        pi = partition(arr, low, high)

        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
  
  
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quickSort(arr, 0, n-1)
print("Sorted array is:")
for i in range(n):
    print("%d" % arr[i])
```

```python
# Heap Sort
def heapify(arr, n, i):
    largest = i  
    l = 2 * i + 1     
    r = 2 * i + 2    
    if l < n and arr[i] < arr[l]:
        largest = l
    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  # swap
        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)
  
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
  
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)


arr = []
for x in range(10):
    a = random.randint(-50, 50)
    arr.append(a)
    
print("Array:", arr)
print('-'*50)
heapSort(arr)
n = len(arr)
print ("Sorted array is")
for i in range(n):
    print ("%d" %arr[i])
```
